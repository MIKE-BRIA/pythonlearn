## Grouping

```python
## Here we will use the (group by) function

# to group a certain column eg Country column

countrygrp=df.groupby(['Country'])
```

```python
# how to look at a group of stuff here by country name
# here now every information comes with that country we specified
countrygrp.get_group('Uganda')
```

```python
# To find information on a particular column like the gender in uganda
# we can first filter


filt = df['Country']=="Uganda"
df.loc[filt]['Gender'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
# To look at the median in a country 

countrygrp['CompTotal'].median()
```

```python
 # to see median salary in a specific country
    
countrygrp['CompTotal'].median().loc['Germany']
```

```python
# To see more aggregate functions(.agg)

countrygrp['CompTotal'].agg(['median','mean']).tail(50)
```

## Working with Dates and Time Series Data

```python
boom = pd.read_csv('Desktop/pandas/ETH_1h.csv')
```

```python
boom
```

```python
boom.head()
```

```python
# To run a date time method
# first lets pick on date that is for the 0 index

boom.loc[0,'Date']
```

```python
boom.dtypes
```

```python
# ton convert the date from a string to a date_time

boom['Date'] = pd.to_datetime(boom['Date'], format= '%Y-%m-%d %I-%p')
```

```python
boom['Date']
```

```python
# to get the day name

boom.loc[0, 'Date'].day_name()
```

```python
# To get the dat name of all the days we will use the date time function(.dt.day_name())

boom['Date'].dt.day_name()
```

```python
# To create a new column  

boom['dayofWeek']=boom['Date'].dt.day_name()
boom
```

```python
# Exploring the data
# viewing the earlyist date and most recent date

boom['Date'].min()
```

```python
boom['Date'].max()
```
