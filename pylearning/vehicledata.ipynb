## Region with the highest purchase rate of used cars

```python
# group by

group=(pom.groupby(['region'])['price'].sum())
group.head(40)
```

```python
gp=(group.reset_index(name="values"))
gp
```

### top 30 regions with high sale of used cars

```python
top=(gp.sort_values("values", ascending=False))
top.head(30)
```

### regions with the least amount of used cars sold

```python
top.tail(30)
```
