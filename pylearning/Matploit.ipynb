# Matpliotlib tutorial

```python
# first import matplotlib

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[11,15,9,7,13,12,5,5,12]
```

```python
# to plot

plt.plot(x,y)
```

```python
# to add color (supply color argument) and if you want the line to be thicker (supply linewidth)

plt.plot(x,y, color="yellow", linewidth=4)
```

```python
# to set x labels , y labels together with tittle

plt.xlabel('month')
plt.ylabel('stocks')
plt.title('Stocks Profit')
plt.plot(x,y, color="green", linewidth=4)
```

### format strings in plot function

## Axes labels, Legend, Grid

```python
days =[1,2,3,4,5,6,7]
max_t=[50,53,45,51,32,40,45]
min_t=[43,42,40,44,43,35,37]
avg_t=[45,47,42,46,37,36,38]
```

```python
# To plot a single one you will()

plt.plot(days,max_t)
```

```python
# But to plot all of them( )

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
# To have a title, label for both the x and y axis

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
# We know need to add legends to our chart so as to identify what are the lines in our chart
# That we will first add labels 



plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t, label="max")
plt.plot(days,min_t, label="min")
plt.plot(days,avg_t, label='avg')

# To place the legend
plt.legend(loc="lower left")

# You can add shadows and fontsize to the legend and even more
```

```python
# To add a grid line


plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t, label="max")
plt.plot(days,min_t, label="min")
plt.plot(days,avg_t, label='avg')

plt.legend(loc="lower left")

# adding grid line
plt.grid()
```

## Bar Chart

```python
# Here we will be required to import numpy because we will have to use it

import numpy as np
```

```python
# The data

company=['BIMAX', 'UNI', 'MAD', 'CHAT']
revenue=[900,1136,809,100]
```

```python
# When ploting bar chart the x axis does not honor string values that's why we will need numpy
# so i will use the xpos

xpos = np.arange(len(company))
xpos

# the arrays you see below are the position of the name of the companies listed
```

```python
plt.bar(xpos,revenue)
```

```python
# To replace the x axis number values by the names of the companies
# We will use (.xticks)

plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
# To add X and Y labels

plt.xticks(xpos,company)
plt.xlabel('companies')
plt.ylabel('revenue(billions)')
plt.title('KENYA TECH STOCKS')
plt.bar(xpos,revenue)
```

```python
# WORKING WITH BOTH REVENUE AND PROFITS FOR THIS COMPANIES

company=['BIMAX', 'UNI', 'MAD', 'CHAT']
revenue=[900,1136,809,600]
profit=[370,420,160,350]
```

```python
# PLOT THE BAR CHART

plt.xticks(xpos,company)
plt.xlabel('companies')
plt.ylabel('revenue(billions)')
plt.title('KENYA TECH STOCKS')
plt.bar(xpos,revenue, label='revenue')
plt.bar(xpos,profit, label='profit')
plt.legend()
```

```python
# TO SEPARATE THE TWO BARS WE WILL HAVE TO DIFFERENTIATE THE XPOS AND ADD A WIDTH FUNCTION

plt.xticks(xpos,company)
plt.xlabel('companies')
plt.ylabel('revenue(billions)')
plt.title('KENYA TECH STOCKS')
plt.bar(xpos-0.2,revenue, width=0.4,label='revenue')
plt.bar(xpos+0.2,profit, width=0.4,label='profit')
plt.legend()
```

```python
# LET'S MAKE THIS BAR CHART TO BE HORIZONTAL

plt.yticks(xpos,company)
plt.title('KENYA TECH STOCKS')
plt.ylabel('companies')
plt.xlabel('revenue(billions)')
plt.barh(xpos-0.2,revenue,height=0.4,label='revenue')
plt.barh(xpos+0.2,profit,height=0.4,label='profit')
plt.legend()
```

## HISTOGRAM

```python
# DATA
# HISTOGRAM ONLY WORK WITH A SINGLE ARRAY OF DATA

blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129 ]
```

```python
# PLOTING A HISTOGRAM
plt.hist(blood_sugar)
```

```python
# THE DEFAULT NUMBER OF CONTAINERS/ BINS =10 LET'S REDUCE THAT TO MAKE OUR HISTOGRAM MUCH BETTER
# BINS IS THE RANGE IN WHICH THE DATA IS DIVIDED INTO

plt.hist(blood_sugar,bins=4)
```

```python
# TO MAKE A SPACING BETWEEN THE BARS (rwidth)

plt.hist(blood_sugar,bins=4, rwidth=0.94)
```

```python
# TO PROVIDE YOUR OWN RANGES FOR THE DATA

plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.94, color='b')
```

```python
# WHEN DEALING WITH MORE THAN ONE DATASET

blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.98, color=['green','orange'], label=['men','women'])
plt.legend()

plt.xlabel('sugar range')
plt.ylabel('number of people')
plt.title('SUGAR TEST')
```

```python
# TO PLOT HISTOGRAM HORIZONTALLY USE THE ORIENTATION FUNCTION


blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.98, color=['green','orange'], 
         label=['men','women'], orientation='horizontal')
plt.legend()

plt.xlabel('sugar range')
plt.ylabel('number of people')
plt.title('SUGAR TEST')
```

## PIE CHART

```python
# DATA

exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/internet Bill","Car","Other Utilities"]
```

```python
# PLOTING

plt.pie(exp_vals,labels=exp_labels)
```

```python
# TO REMOVE THE EXTRA LABELS AND TO MAKE SURE ALL YOUR AXIS ARE EQUAL

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
# TO INCREASE THE SIZE USE ( RADIUS) ARGUMENT BY DEFAULT IT IS ONE

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=2)
plt.show()
```

```python
# TO ADD PERCENTAGE OF THE PIES WITH 2 DECIMAL PLACES

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.2f%%')
plt.show()
```

```python
# TO TAKE OUT A PIE (EXPLODE) SPECIFY NUMBER OF VALUES

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.2f%%', explode=[0,0.2,0,0,0.1],shadow=True)
plt.show()
```
